```python
losses = []
i = 0
loss = []
for line in open('./log/Model_Sat Sep 10 22:44:20 2022.log','r'):
    i += 1
    if i % 2 == 1:
        continue
    tmp = line.split(' ')
    loss.append(float(tmp[-1][:-2]))
losses.append(loss[0::2])
i = 0
loss = []
for line in open('./log/Model_2022-09-21_16:55:43.log','r'):
    i += 1
    if i % 2 == 1:
        continue
    tmp = line.split(' ')
    loss.append(float(tmp[-1][:-2]))
losses.append(loss)
i = 0
loss = []
for line in open('./log/Model_Sat Sep 10 20:04:37 2022.log','r'):
    i += 1
    if i % 2 == 1:
        continue
    tmp = line.split(' ')
    loss.append(float(tmp[-1][:-2]))
losses.append(loss[:-1])
i = 0
loss = []
for line in open('./log/Model_2022-09-23_14:11:00.log','r'):
    i += 1
    if i % 2 == 1:
        continue
    tmp = line.split(' ')
    loss.append(float(tmp[-1][:-2]))
loss+=[0.337281,0.320212,0.318847,0.315672,0.312331]
losses.append(loss)
i = 0
loss = []
for line in open('./log/Model_Tue Sep 06 22:20:03 2022.log','r'):
    i += 1
    if i == 1:
        continue
    tmp = line.split(' ')
    loss.append(float(tmp[-1][:-2]))
losses.append(loss[0::2])
```

```python
from matplotlib.ticker import MultipleLocator, FormatStrFormatter, FuncFormatter
x_tick = [32,64,128,256,512,1024]
y_tick_mrr = [0.398,0.513,0.598,0.612,0.622,0.629]
y_tick_hit3 = [0.424,0.56,0.652,0.668,0.681,0.691]
y_tick_hit5 = [0.488,0.644,0.734,0.743,0.751,0.757]
y_tick_hit10 = [0.564,0.726,0.812,0.823,0.835,0.841]
x_1 = [x+0.5 for x in range(len(x_tick))]
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.plot(x_1,y_tick_mrr,linestyle='-',color='orange',marker='^',label='MRR')
ax.plot(x_1,y_tick_hit3,linestyle='-',color='limegreen',marker='o',label='HIT@3')
ax.plot(x_1,y_tick_hit5,linestyle='-',color='royalblue',marker='x',label='HIT@5')
ax.plot(x_1,y_tick_hit10,linestyle='-',color='gold',marker='P',label='HIT@10')
plt.xticks([x for x in range(len(x_tick))])
def get_dimension(x,pos):
    # dimensions = [str(x) for x in x_tick]
    # print(x)
    return str(int(2**(5+x-0.5)))
xminorLocator = MultipleLocator(0.5)
ax.xaxis.set_minor_locator(xminorLocator)
ax.xaxis.set_minor_formatter(FuncFormatter(get_dimension))
plt.xlabel('Dimension')
plt.ylabel('MRR&HIT@K')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# for a,b in zip(x_1,y_tick):
#     plt.text(a,b+0.002,'%.3f'%b,ha='center',va='bottom',fontsize=10)
plt.grid(visible=True,axis='x',which='major',linestyle='--',color='gainsboro',linewidth=0.75)
ax.xaxis.set_major_formatter(plt.NullFormatter())
# ax.xaxis.set_major_locator(plt.NullLocator())
ax.tick_params(axis='x',which='major',color='white')
plt.legend()
```
